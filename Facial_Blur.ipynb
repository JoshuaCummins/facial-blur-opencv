```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
ww = cv2.imread('ww.jpg')
ww = cv2.cvtColor(ww, cv2.COLOR_BGR2RGB)
```

```
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```
def display(img):
  fig = plt.figure(figsize=(10,10))
  ax = fig.add_subplot(111)
  #new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  ax.imshow(img)
```

```
def detect_face(img):
    
  
    face_img = img.copy()
  
    face_rects = face_cascade.detectMultiScale(face_img) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
    
```

```
result = detect_face(ww)
```

```
display(result)
plt.title('face detected')
plt.axis('off')
```

```
def detect_and_blur_face(img):
    
    face_img = img.copy()
    roi = img.copy()

    rectangles = face_cascade.detectMultiScale(face_img, scaleFactor=1.3, minNeighbors=3)
    for (x,y,w,h) in rectangles:

      roi = roi[y:y+h, x:x+w]
      blurred_roi = cv2.medianBlur(roi, 9)

      face_img[y:y+h, x:x+w] = blurred_roi
    
    return face_img
```

```
result = detect_and_blur_face(ww)
```

```
display(result)
plt.title('blurred face')
plt.axis('off')
```
